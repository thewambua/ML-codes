```python
#Importint libraries and the dataset
import pandas as pd 

data = pd.read_csv('machinesensordata.csv')
data
```

```python
data.info()
data.describe()
```

```python
#checking for null values
null_values = data.isnull().sum()
null_values
```

```python
#label encoding
#Label encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in data.columns:
    if data[col].dtype == object:
        data[col] = le.fit_transform(data[col])
data
```

```python
#Splitting data into X and y
X = data.drop(['Machine failure', 'TWF', 'HDF', 'PWF', 'OSF', 'RNF'], axis=1)
X
```

```python
y = data['Machine failure']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)

X_train
```

```python
X_test
```

```python
#MODEL
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
y_train_pred
```

```python
y_test_pred
```

```python
accuracy = r2_score(y_test, y_test_pred)
print(f"R-squared accuracy: {accuracy:.2f}")
```
